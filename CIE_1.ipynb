```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
d=pd.read_csv('mtcars.csv')
d
```

```python
plt.title("HISTOGRAM")
hist=sns.histplot(x='mpg',data=d,bins=10,kde=False)
hist
```

```python
plt.title("SCATTERPLOT")
scatter=sns.scatterplot(x='wt',y='mpg',hue='mpg',data=d)
```

```python
bar=sns.countplot(x='am',data=d)
```

```python
box=sns.boxplot(x='mpg',data=d)
box
```

```python
import pandas as pd
data=pd.DataFrame({'ProductName':['washingmachine','toaster','bulb','kidsware','dresses'],
               'category':['home','kithen','home','cloth','cloth'],
               'price':[17000,3000,400,3000,2300],
               'Quantity':[1,3,5,3,7]})
data
```

```python
p=data.pivot_table(index=['category'],values=['price'],aggfunc='mean')
p
```

```python
data['revenue']=data['price']*data['Quantity']

p=data.pivot_table(index=['category'],values=['price','revenue'],aggfunc='mean')
p
```
